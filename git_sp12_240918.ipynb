# Описание проекта

HR-аналитики компании «Работа с заботой» помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.
    
Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 
    
Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли. 
    
У вас будет несколько задач. Первая — построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика. 
    
Почему бизнесу это важно: удовлетворённость работой напрямую влияет на отток сотрудников. А предсказание оттока — одна из важнейших задач HR-аналитиков. Внезапные увольнения несут в себе риски для компании, особенно если уходит важный сотрудник.
    
Ваша вторая задача — построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.
    
Теперь расскажем подробнее о задачах

```python
# Функция построения графиков и гистограмм датафрейма

# Логика работы функции:
# - для числовых признаков: если признак дискретный то строим sns.countplot, 
#       иначе hist и sns.boxplot;
# - для остальных (категориальных) - plt.pie

def graph_col(x): 
    
    for each in x.columns: # Обработка числовых признаков
        if (x[each].dtype != 'object'):
            k = x.columns.get_loc(each)
            
            if ('int' not in str(x[each].dtype)) or (x[each].value_counts().count() > 10): # Обработка недискретных признаков
                
                fig, axs = plt.subplots(1, 2, figsize = (12, 4))
                plt.subplots_adjust(wspace = 1)
                
                sns.boxplot(x[each], ax = axs[0]);
                axs[0].set_title(f'Диаграмма размаха значений {each}')
            
                axs[1] = x[each].hist(bins = 50)
                plt.title(f'График распределения количества сотрудников по параметру {each}')
                plt.xlabel(each)
                plt.ylabel('Количество сотрудников')
                
            else: # Обработка дискретных признаков
                
                fig, ax = plt.subplots()
                sns.countplot(x[each]); 
                ax.set_ylabel('Количество сотрудников')
                plt.suptitle(f'График распределения количества сотрудников по параметру {each}')
                
                # Рисование процентов
                total = x[each].count()

                for p in ax.patches:
                    xc = p.get_x() + p.get_width() / 2 - 0.05
                    yc = p.get_y() + p.get_height()
                    ax.annotate(f'{p.get_height()/total * 100:.2f}%', (xc - 0.3 , yc + total/500), size = 10)   
             
            plt.show()   
            
    for each in x.columns: # Обработка категориальных признаков
        if x[each].dtype == 'object':
            plt.title(f'График распределения количества сотрудников по параметру {each}')
            plt.pie(
                x.pivot_table(index = each, values = x.columns[k], aggfunc = 'count')[x.columns[k]]
                , labels = x.pivot_table(index = each, values = x.columns[k], aggfunc = 'count').index
                , autopct = '%1.1f%%'
                )
            #plt.title(each) 
            plt.show()
```

## Шаг 1.4. Подготовка данных

Создаем тренировочную и тестовую выборки. Целевой признак - 'job_satisfaction_rate'

```python
RANDOM_STATE = 42

X_train = dfs_dict['train_job_satisfaction_rate'].drop(['id', 'job_satisfaction_rate'], axis = 1)
X_test = dfs_dict['test_features'].drop(['id', 'job_satisfaction_rate'], axis = 1)
y_train = dfs_dict['train_job_satisfaction_rate']['job_satisfaction_rate']
y_test = dfs_dict['test_features']['job_satisfaction_rate']

#display(X_train.head())
```

Создаем список числовых признаков и два списка категориальных (для кодирования с помощью OneHotEncoder и OrdinalEncoder)

Признак 'dept' будем кодировать с помощью OneHotEncoder (у 'dept' больше всего значений среди категориальных признаков, и они не образуют иерархию), остальные категориальные - с помощью OrdinalEncoder

```python
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']
ohe_columns = ['dept']
ord_columns = ['level', 'workload', 'last_year_promo', 'last_year_violations']
```

Пайплайн SimpleImputer + OneHotEncoder

```python
ohe_pipe = Pipeline(
    [('simpleImputer', impute_pipe),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )
```

Пайплайн SimpleImputer + OrdinalEncoder

```python
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', impute_pipe),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'], # 'level'
                    ['low', 'medium', 'high'], # 'workload'
                    ['no', 'yes'], # 'last_year_promo'
                    ['no', 'yes']  # 'last_year_violations'
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', impute_pipe)
    ]
)
```

Общий пайплайн для подготовки данных

```python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough',
)
```

Итоговый пайплайн: подготовка данных и параметры

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])
```

## Шаг 1.5. Обучение моделей

Создадим свою метрику SMAPE

```python
def SMAPE(y_true, y_pred):
    res = 100 * (abs(y_true - y_pred) / ((abs(y_true) + abs(y_pred)) / 2)).sum() / len(y_true)
    #print(res)
    return res
```

```python
my_scorer_SMAPE = make_scorer(SMAPE, greater_is_better = False)
```

Проверяем гипотезу: Средние значения job_satisfaction_rate у уволившихся и оставшихся сотрудников равны.

Используем двухвыборочный t-тест для независимых выборок, st.ttest_ind

Первая выборка - записи test_quit['job_satisfaction_rate'] где test_quit['quit'] = 'no'

Вторая выборка - записи test_quit['job_satisfaction_rate'] где test_quit['quit'] = 'yes'

Нулевая гипотеза - Средние значения job_satisfaction_rate у уволившихся и оставшихся сотрудников равны

Альтернативная гипотеза - Средние значения job_satisfaction_rate у уволившихся и оставшихся сотрудников не равны
    
Статистическая значимость = 0.05

```python
alpha = 0.05 # Выбираем статистическую значимость

results = scipy.stats.ttest_ind(test_quit[test_quit['quit'] == 'no']['job_satisfaction_rate'], 
          test_quit[test_quit['quit'] == 'yes']['job_satisfaction_rate'], 
          equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print ('Отвергаем нулевую гипотезу. Средние значения job_satisfaction_rate у уволившихся и оставшихся сотрудников не равны')
else:
    print ('Не отвергаем нулевую гипотезу. ' 
           'С большой долей вероятности Средние значения job_satisfaction_rate у уволившихся и оставшихся сотрудников '
           'не имеют статистически значимых отличий')

```

В качестве значения ценности используем признак salary (в принципе можно много различных критериев ценности сочинить). Т.е. в первую очередь в нашем случае интересны высокооплачиваемые сотрудники
